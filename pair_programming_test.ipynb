```python
import math

def polar_to_cartesian(radius, angle_radians):
    # x = r * cos(theta), y = r * sin(theta)
    x = radius * math.cos(angle_radians)
    y = radius * math.sin(angle_radians)
    return x, y

# Test 1: Radius 5, Angle 0 (Should be 5.0, 0.0)
print(f"Test 1: {polar_to_cartesian(5, 0)}")

# Test 2: Radius 10, Angle pi/2 (Should be ~0, 10.0)
print(f"Test 2: {polar_to_cartesian(10, math.pi/2)}")
```
