```python
from pair_programming import convert_to_meters
import math
```

```python
# Test 1: A standard height (e.g., 6 feet 0 inches)
feet1, inches1 = 6, 0
meters1 = convert_to_meters(feet1, inches1)

print(f"Test 1 Output: {meters1} meters")

# 6 feet is exactly 72 inches. 72 * 0.0254 = 1.8288
assert math.isclose(meters1, 1.8288), "Should be exactly 1.8288 meters"
print("Test 1 Passed!")
```

```python
# Test 2: An edge case using only inches (0 feet, 12 inches)
feet2, inches2 = 0, 12
meters2 = convert_to_meters(feet2, inches2)

print(f"Test 2 Output: {meters2} meters")

# 12 inches is exactly 0.3048 meters
assert math.isclose(meters2, 0.3048), "Should be exactly 0.3048 meters"
print("Test 2 Passed!")
```
